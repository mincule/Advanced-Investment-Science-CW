## 2.2.1. Data preprocessing
Before we select the stocks, we need to preprocess s&p stock data. It is because we will do historical simulation for independent 3 times and we do not use all of the s&p500 stocks. 

###### Functions Contents

- dataprocessing_1
- dataprocessing_2

```python
def dataprocessing_1(df, start_date, end_date):
    """it makes a mask."""

    mask = (df.index >= pd.to_datetime(start_date)) & (df.index <= pd.to_datetime(end_date))
    df_masked = df.loc[mask]
    
    return df_masked


def dataprocessing_2(df_masked):
    """
        it deletes symbols which do not have price data "at the Quarter".
    """
    no_history = []
    symbols_df_masked = df_masked.columns.tolist()
    for sym in symbols_df_masked:
        statement = (pd.isnull(df_masked[sym][0]) and pd.isnull(df_masked[sym][1]) and pd.isnull(df_masked[sym][2])
                     and pd.isnull(df_masked[sym][3]) and pd.isnull(df_masked[sym][4]) and pd.isnull(df_masked[sym][5]))
        if statement:
            df_masked = df_masked.drop([sym], axis=1)
            no_history.append(sym)
  
    return df_masked, no_history
```

## 2.2.2. Accelerating Dual Momentum (ADM) method

The "Accelerating Dual Momentum (ADM)" is one of the tactical asset allocation strategy. This strategy use momentum scores of calculating sum = 1-month, 3-month and 6-month returns. In general, asset classes which have more positive scores are chosen and rebalance once per month.

Here, we will use ADM scores to select the TOP 10 stocks. The reason restricting stocks to 10 is that managing is easy. And rebalace every quarter.

###### Functions Contents

- get_ADM
- Asset_selection
- df_sel

```python
# Accelaerating Dual Momentum (ADM)
def get_ADM(df, sym):
    date_initial = ''
    date_1m = ''
    date_3m = ''
    date_6m = ''
    
    
    for i in range(6):
        if pd.notnull(df[sym][i]):
            date_initial = df[sym].index[i]
            break
    for i in range(6):
        if pd.notnull(df[sym][i+19]):
            date_1m = df[sym].index[i+19]
            break
    for i in range(6):
        if pd.notnull(df[sym][i+39]):
            date_3m = df[sym].index[i+39]
            break
    for i in range(6):
        if pd.notnull(df[sym][i+59]):
            date_6m = df[sym].index[i+59]
            break
    
    #print(sym, date_initial, date_1m, date_3m, date_6m)
    momentum_1 = (df[sym][date_1m] - df[sym][date_initial]) / df[sym][date_initial]
    momentum_3 = (df[sym][date_3m] - df[sym][date_initial]) / df[sym][date_initial]
    momentum_6 = (df[sym][date_6m] - df[sym][date_initial]) / df[sym][date_initial]
    score = (momentum_1 + momentum_3 + momentum_6) / 3
    
    return score


def Asset_selection(df_masked):
    ADMscore = {}
    selected = []
    symbols_df_masked = df_masked.columns.tolist()
    
    for sym in symbols_df_masked:
        ADMscore[sym] = get_ADM(df_masked, sym)
    ADMscore = sorted(ADMscore.items(), key=lambda x: x[1], reverse=True)
    
    for sel in range(10):
        selected.append(ADMscore[sel])
        
    return selected


def df_sel(df_masked, selected):
    symbols_sel = []
    df_invest = pd.DataFrame()
    
    for i in range(len(selected)):
        symbols_sel.append(selected[i][0])

    for sym in symbols_sel:
        df_invest = pd.concat([df_invest, df_masked[sym]], axis=1)
        
    return df_invest
```

### 2.3.1. Hierarchical Risk Parity (HRP) Portfolio
Hierarchical Risk Parity portfolio is one of the risk-parity portfolios which constructs diversified portfolio without considering return correlation of assets in other clusters. HRP makes clusters and so assets have hierarchy. This makes it possible to compete (weight allocation) with each other within the group(cluster). So, we can diversify well.<br>

We use HRP Portfolio for Housing investment because it has top-down fashion, more intuitive and manageable.

HRP algorithm consists of 3 parts:<br>
1. Hierarchical Clustering
2. Quasi-Diagonalisation
3. Recursive Bisection

Time complexity is known as O(n^3).

Reference:
 - https://hudsonthames.org/an-introduction-to-the-hierarchical-risk-parity-algorithm/
 - https://github.com/TheRockXu/Hierarchical-Risk-Parity/blob/master/Hierarchical%20Clustering.ipynb

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from warnings import simplefilter
simplefilter("ignore", sch.ClusterWarning)
```

### Hierarchical Clustering
We will perform Hierarchical clustering with a distance matrix of correlation.<br>

Repeat the procedure until all of the clusters are merged together:
   1. Identify 2 clusters which are closest.
   2. Merge them.
    
a distance metric: ((1-corr)/2.)**.5

```python
def correlDist(corr):
    dist = ((1-corr)/2.)**.5 # distance matrix
    
    return dist

def get_link(df_adm_period):
    corr = df_adm_period.corr()
    dist = correlDist(corr)
    link = sch.linkage(dist, 'single')
    df_link = pd.DataFrame(link)
    
    return df_link, link
```

### Quasi-Diagonalisation
a kind of a matrix seriation <br> 
Rearrange the rows and columns of the covariance matrix of stocks. This makes similar investments stick together or far apart.

```python
def get_quasi_diag(link):
    
    link = link.astype(int)
    
    # get the first and the second item of the last tuple
    sort_ix = pd.Series([link[-1,0], link[-1,1]]) 
    
    # the total num of items is the third item of the last list
    num_items = link[-1, 3] # the total number of stocks
    
    # if the max of sort_ix is bigger than or equal to the max_items
    while sort_ix.max() >= num_items: # this means that ....
        
        sort_ix.index = range(0, sort_ix.shape[0]*2, 2) # indexing with even numbers
        
        df0 = sort_ix[sort_ix >= num_items] # find clusters with even number indexing
        
        # df0 contain even index and cluster index
        i = df0.index
        j = df0.values - num_items
        
        sort_ix[i] = link[j,0] # item 1
        
        df0  = pd.Series(link[j, 1], index=i+1)
        
        sort_ix = sort_ix.append(df0)
        sort_ix = sort_ix.sort_index()
        
        sort_ix.index = range(sort_ix.shape[0])
        
    
    return sort_ix.tolist()
```

### Recursive Bisection
Allocate the weights to each cluster.<br>
This shows the only assets within the same group compete for allocation with each other rather than competing all the assets in the portfolio.

 

```python
def get_cluster_var(cov, c_items):
    cov_ = cov.iloc[c_items, c_items] # matrix slice
    # calculate the inversev-variance portfolio
    ivp = 1./np.diag(cov_)
    ivp/=ivp.sum()
    w_ = ivp.reshape(-1,1)
    c_var = np.dot(np.dot(w_.T, cov_), w_)[0,0]
    return c_var


def get_rec_bipart(cov, sort_ix):
    # compute HRP allocation
    # intialize weights of 1
    w = pd.Series(1, index=sort_ix)
    
    # intialize all items in one cluster
    c_items = [sort_ix]
    while len(c_items) > 0:
        # bisection
        c_items = [i[int(j):int(k)] for i in c_items for j,k in 
                   ((0,len(i)/2),(len(i)/2,len(i))) if len(i)>1]
        
        # now it has 2
        for i in range(0, len(c_items), 2):
            
            c_items0 = c_items[i] # cluster 1
            c_items1 = c_items[i+1] # cluter 2
            
            c_var0 = get_cluster_var(cov, c_items0)
            c_var1 = get_cluster_var(cov, c_items1)
            
            alpha = 1 - c_var0/(c_var0+c_var1)
            
            w[c_items0] *= alpha
            w[c_items1] *=1-alpha
    return w
```

Using below function when construct HRP portfolios.

```python
def get_HRP(df_history):
    # get log returns
    df_log = np.log(df_history / df_history.shift(1))
    df_log = df_log.fillna(0.0)
    
    # hierarchical clustering
    df_link, link = get_link(df_log)
    dendro = sch.dendrogram(df_link)
    
    # quasi-diag.
    sort_ix = get_quasi_diag(link)
    
    # recursive bisection
    cov = df_log.cov()
    weights = get_rec_bipart(cov, sort_ix)
    
    re = []
    for i in range(len(df_history.columns)):
        re.append(i)
    weights = weights.reindex(re)
    
    return weights
```

### 2.3.2. MVO Portfolio

###### Functions Contents

- get_MVO

```python
import cvxpy as cp
```

```python
def get_MVO(df_history):
    
    # get log returns
    df_log = np.log(df_history / df_history.shift(1))
    df_log = df_log.fillna(0.0)
    
    # MVO
    mu = (df_log.mean() * len(df_log)).to_numpy()
    Sigma = (df_log.cov() * len(df_log)).to_numpy()
    
    w = cp.Variable(len(df_history.columns))
    gamma = 0.01
    
    ret = mu.T @ w
    risk = cp.quad_form(w, Sigma)
    prob = cp.Problem(cp.Minimize(risk-gamma*ret),
                     [cp.sum(w) == 1,
                     w >= 0])

    prob.solve()
    
    return w.value
```

### 2.3.3. Rebalancing

Rebalance every quarter. For 2 years investment, we should construct total 8 portfolios.<br>


An investment starts in 07.01. and ends in 06.30.

ex) start year 2015 investment is: 2015.07.01. ~ 2017.06.30.

###### Functions Contents

- investment_HRP
- investment_MVO
- investment_whole
- investment_result

```python
def investment_HRP(df, start_date):
    selected = []
    weights = []
    
    # total 7 phases (for calculating ADM and resampling assets)
    phase = []
    phase0 = ['-01-01','-06-30']
    phase1 = ['-04-01','-09-30']
    phase2 = ['-07-01','-12-31']
    phase3 = ['-10-01','-03-31']
    phase.append(phase0)
    phase.append(phase1)
    phase.append(phase2)
    phase.append(phase3)
    
    for i in range(8):
        if i == 0 or i == 1 or i == 2 or i == 3:
            if i == 0 or i == 1 or i == 2:
                s_date = str(start_date) + phase[i][0]
                e_date = str(start_date) + phase[i][1]
            else:
                s_date = str(start_date) + phase[i][0]
                e_date = str(start_date + 1) + phase[i][1]
                
        else:
            if i == 4 or i == 5 or i == 6:
                s_date = str(start_date + 1) + phase[i-4][0]
                e_date = str(start_date + 1) + phase[i-4][1]
            else:
                s_date = str(start_date + 1) + phase[i-4][0]
                e_date = str(start_date + 2) + phase[i-4][1]
        
        df_phase = dataprocessing_1(df, s_date, e_date)
        
        df_phase, no_history  = dataprocessing_2(df_phase)
        selected_assets = Asset_selection(df_phase)
        df_invest = df_sel(df_phase, selected_assets)
        w = get_HRP(df_invest)
        
        selected.append(df_invest.columns.to_list())
        weights.append(w)
    
    return selected, weights, start_date


def investment_MVO(df, start_date):
    selected = []
    weights = []
    
    # total 7 phases (for calculating ADM and resampling assets)
    phase = []
    phase0 = ['-01-01','-06-30']
    phase1 = ['-04-01','-09-30']
    phase2 = ['-07-01','-12-31']
    phase3 = ['-10-01','-03-31']
    phase.append(phase0)
    phase.append(phase1)
    phase.append(phase2)
    phase.append(phase3)
    
    for i in range(8):
        if i == 0 or i == 1 or i == 2 or i == 3:
            if i == 0 or i == 1 or i == 2:
                s_date = str(start_date) + phase[i][0]
                e_date = str(start_date) + phase[i][1]
            else:
                s_date = str(start_date) + phase[i][0]
                e_date = str(start_date + 1) + phase[i][1]
                
        else:
            if i == 4 or i == 5 or i == 6:
                s_date = str(start_date + 1) + phase[i-4][0]
                e_date = str(start_date + 1) + phase[i-4][1]
            else:
                s_date = str(start_date + 1) + phase[i-4][0]
                e_date = str(start_date + 2) + phase[i-4][1]
        
        df_phase = dataprocessing_1(df, s_date, e_date)
        
        # print("phase  ", i)
        # print(s_date, e_date)
        # print(df_phase)
        
        df_phase, no_history  = dataprocessing_2(df_phase)
        selected_assets = Asset_selection(df_phase)
        df_invest = df_sel(df_phase, selected_assets)
        w = get_MVO(df_invest)
        
        selected.append(df_invest.columns.to_list())
        weights.append(w)
    
    return selected, weights, start_date
```

```python
def investment_result(df_invest, initial_outlay, weights, PORTFOLIO_NAME):
    dollar = np.zeros(len(df_invest)+1)
    returns = np.zeros(len(df_invest))
    dollar[0] = initial_outlay

    for i in range(len(df_invest)):
        dollar[i+1] = round(dollar[i] * np.exp(df_invest.iloc[i].to_numpy().dot(weights)),0)
        returns[i] = dollar[i+1]
        
    return returns


def investment_whole(PORTFOLIO_NAME, initial_outlay, start_date, selected, weights):
    returns_whole = []
    rr=[]
    date = []
    
    # total 7 phases (for calculating ADM and resampling assets)
    phase = []
    phase0 = ['-07-01','-09-30']
    phase1 = ['-10-01','-12-31']
    phase2 = ['-01-01','-03-31']
    phase3 = ['-04-01','-06-30']
    phase.append(phase0)
    phase.append(phase1)
    phase.append(phase2)
    phase.append(phase3)
    
    for i in range(8):
        if i == 0 or i == 1:
            s_date = str(start_date) + phase[i][0]
            e_date = str(start_date) + phase[i][1]
        elif i == 2 or i == 3 or i == 4 or i == 5:
            if i == 2 or i ==3:
                s_date = str(start_date + 1) + phase[i][0]
                e_date = str(start_date + 1) + phase[i][1]
            else:
                s_date = str(start_date + 1) + phase[i-4][0]
                e_date = str(start_date + 1) + phase[i-4][1]
                
        elif i == 6 or i == 7:
            s_date = str(start_date + 2) + phase[i-4][0]
            e_date = str(start_date + 2) + phase[i-4][1]
        
        df_phase = dataprocessing_1(df, s_date, e_date)
        
        temp = []
        for j in range(len(selected[i])):
            temp.append([selected[i][j], -999])
        
        df_invest = df_sel(df_phase, temp)
        df_invest = np.log(df_invest / df_invest.shift(1))
        df_invest = df_invest.fillna(0.0)
        
        if i == 0:
            outlay = initial_outlay
        else:
            outlay = returns_whole[i-1][-1]
        
        w = weights[i]
        returns = investment_result(df_invest, outlay, w, PORTFOLIO_NAME)
        returns_whole.append(returns)
        date.extend(df_invest.index.to_list())
        
    for i in range(len(returns_whole)):
        tt = list(returns_whole[i])
        rr.extend(tt)
    
    # get visual
    import matplotlib.pyplot as plt
    fig = plt.figure()
    fig.add_subplot(1, 1, 1)
    if PORTFOLIO_NAME == "HRP":
        plt.plot(date, rr, color='#23c8c4')
    elif PORTFOLIO_NAME == "MVO":
        plt.plot(date, rr, color='#dc373b')
    plt.xticks(rotation = 45)
    plt.ylabel("account value [USD]")
    plt.xlabel("date")
    plt.text(date[-1],rr[-1],
         rr[-1],
         color='#003458',
         horizontalalignment='center',
         verticalalignment='top')
    plt.title(str(PORTFOLIO_NAME) + " portfolio investment")
    plt.grid(True)
    plt.show()
    
    return rr
```

```python
def get_log_return(df_masked):
    df_masked = np.log(df_masked / df_masked.shift(1))
    df_masked = df_masked.fillna(0.0)
    
    return df_masked


def sp_result(sp, initial_outlay):
    dollar = np.zeros(len(sp)+1)
    returns = np.zeros(len(sp))
    dollar[0] = initial_outlay

    for i in range(len(sp)):
        dollar[i+1] = round(dollar[i] * np.exp(sp.iloc[i]),0)
        returns[i] = dollar[i+1]
    
    return returns
```

```python
def Sharpe_r(returns):
    r_ratio = []
    for i in range(len(returns)):
        if i == 0:
            pass
        else:
            r_ratio.append((returns[i] - 50000) / 50000)

    r_ratio = pd.Series(r_ratio)
    r_ratio = get_log_return(r_ratio)
    
    mean = pd.Series(r_ratio).mean()
    std = pd.Series(r_ratio).std()
    sharpe_ratio = mean / std
    
    return sharpe_ratio
```

```python
for j in range(len(ww)):
    for i in range(len(ww[j])):
        ratio = ww[j][i]
        labels = ss[j][i]
        colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0', '#D2FFF2', '#258CFF', '#A4CFFF', '#FFD4A4', '#AFA4FF', '#C8FFA3']
        wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

        #plt.subplot(len(ww_2015), 1, 1)
        if j == 0:
            year = 2015
        elif j == 1:
            year = 2017
        elif j == 2:
            year = 2019
            
        plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
        plt.title(str(year) + ' HRP Portfolio' + ' - phase ' + str(i))
        plt.show()
```

Let minus weights (very small weights) be 0 to draw pie charts.

```python
for h in range(len(zz)):
    for j in range(len(zz[h])):
        for i in range(len(zz[h][j])):
            if zz[h][j][i] <= 0 :
                zz[h][j][i] = 0
```

```python
for j in range(len(zz)):
    for i in range(len(zz[j])):
        ratio = zz[j][i]
        labels = zz[j][i]
        colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0', '#D2FFF2', '#258CFF', '#A4CFFF', '#FFD4A4', '#AFA4FF', '#C8FFA3']
        wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

        #plt.subplot(len(ww_2015), 1, 1)
        if j == 0:
            year = 2015
        elif j == 1:
            year = 2017
        elif j == 2:
            year = 2019
            
        plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
        plt.title(str(year) + ' MVO Portfolio' + ' - phase ' + str(i))
        plt.show()
```
